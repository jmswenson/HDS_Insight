```python
import glob
import pandas as pd
```

```python
def z_score_on_pds_cols(pds_df, list_of_cols_to_exclude = ["Width","lineage",\
                                                           "cell_label","screen","well","cell_plate","Time"]):
    '''Calculates the z-score for each column in a pandas dataframe ("pds_df") except those excluded 
        in the list "list_of_cols_to_exclude". Currently it does not handle Nans or missing values intelligently'''
    cols = list(pds_df.columns)
    print(len(cols))
    # Select all columns that you want to do a z-score transform on
    cols_of_int = [x for x in list(pds_df.columns.values) if x not in list_of_cols_to_exclude]

    print(len(cols_of_int))
    pds_df[cols_of_int].head(n=2)
    for col in cols_of_int:
        col_zscore = col + '_zscore'
        pds_df[col_zscore] = (pds_df[col] - pds_df[col].mean())/pds_df[col].std(ddof=0)
    #pds_df.info()
    return pds_df
```
